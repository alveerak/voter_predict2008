```python
import numpy as np

train = np.genfromtxt('train_2008.csv', delimiter=',')
X_train = train[1:, :-1]
y_train = train[1:, -1]
X_test = np.genfromtxt('test_2008.csv', delimiter=',')[1:,:]
```

```python
from sklearn.isotonic import IsotonicRegression
```

```python
clf2 = IsotonicRegression()
clf2.fit_transform(X_train, y_train)
y_test = clf2.predict(X_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

# regularization strengths
a = np.linspace(0.01, 3, 30)
weights = []
# Train linear regression model on data for the given regularization 
# strengths and obtain weights produced from each
for i in range(len(a)):
    clf = Lasso(alpha=a[i])
    clf.fit(X_train, y_train)
    w = clf.coef_
    weights.append(w)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
clf = LogisticRegression(solver='sag',max_iter=5000)
clf.fit(X_train, y_train)
y_test = clf.predict_proba(X_test)
```

```python
output = []
output.append(["id", "target"])
for i in range(len(y_test)):
    #output.append([i, y_predict[i][0]])
    output.append([i, y_test[i][1]])
    
import csv
with open('voter_pred_daniel.csv', mode='w') as voter_file:
    voter_writer = csv.writer(voter_file, delimiter=",")
    for i in output:
        voter_writer.writerow(i)
```
